```python
from SPARQLWrapper import SPARQLWrapper, JSON, CSV

sparql = SPARQLWrapper("http://dbpedia.org/sparql")
docs=0

#Introduce path to save the documents
dir_name= ""
#Introduce a list of Wikipedia categories (eg: Nature) whose documents you want to scrap
categories= [] 

for category in categories:
    
    print("Category")
    print(category)
    
    #SPARQL query to obtain English (language can be modified in the query) abstract tagged with the Wikipedia category noted before
    sparql.setQuery("""
        PREFIX dcterms: <http://purl.org/dc/terms/> 
        prefix dbpedia-owl: <http://dbpedia.org/ontology/>

        SELECT DISTINCT ?resource, ?abstract WHERE {
          ?resource dcterms:subject ?cat .
                ?cat ?p <http://dbpedia.org/resource/Category:"""+category+"""> .
          ?resource dbpedia-owl:abstract ?abstract .
                      filter langMatches(lang(?abstract),"en")

        }
        LIMIT 10000 OFFSET 0
    """)
    sparql.setReturnFormat(JSON)
    results = sparql.query().convert()

    #Abstract and title of the Wikipedia article is obtained in order to create a .txt file
    for result in results["results"]["bindings"]:
        content=result["abstract"]["value"]
        title=result["resource"]["value"]
        title=title[len("http://dbpedia.org/resource/"):]
        if "/" not in title and "List" not in title:
            print("Downloading....")
            print(title)
            text_file = open(dir_name+title+".txt", "wb")
            text_file.write(content.encode('utf-8'))
            text_file.close()
```
